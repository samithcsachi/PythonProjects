```python
from bs4 import BeautifulSoup 
import requests 
```

```python
url = 'https://trendlyne.com/stock-screeners/fundamentals/ROCE_A/roce-annual/highest/index/NIFTY500/nifty-500/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all ('table')[0]
```

```python
soup.find('table', class_ = 'cen alphanum mW140 sorting' )
```

```python
table = soup.find_all ('table')[0]
```

```python
print(table)
```

```python
stock_titles =table.find_all ('th')
```

```python
stock_titles
```

```python
stock_table_titles = [title.text.strip() for title in stock_titles]
print(stock_table_titles)
```

```python
import pandas as pd
```

```python
df =pd.DataFrame(columns=stock_table_titles )
df
```

```python
column_data = table.find_all ('tr')
```

```python
for row in column_data [1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
